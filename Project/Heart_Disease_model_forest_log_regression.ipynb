```python
df2 = df.drop(columns=['HeartDisease', 'Smoking', 'AlcoholDrinking', 'Stroke', 'DiffWalking', 'Sex',
                 'Race', 'Diabetic', 'PhysicalActivity', 'GenHealth','Asthma','KidneyDisease','SkinCancer','AgeCategory'])
```

```python
df2.head()
```

```python
dfs = [df2, smoking,heartdisease,alcoholdrinking ,stroke,diffwalking,
         physicalactivity ,diabetic ,asthma ,skincancer ,sex ,genhealth ,age ,race]
```

```python
dff = pd.concat(dfs, axis=1)
dff.head()
```

```python
list(dff)
```

```python
# Split the data into feature and target data
X = pd.DataFrame(dff.drop(['HeartDisease'], axis = 1))

y = pd.array(dff['HeartDisease'])
```

```python
# Split the data into X_train, X_test, y_train, y_test
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=444)
X_train
```

The Random Forest model was optimized to achieve a high accuracy for both training and test data, both of which are over 90%.

```python
# Train a Random Forest Classifier model and print the model score
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(criterion = "entropy")

forest.fit(X_train, y_train)
predictions = forest.predict(X_test)
y_pred_forest = forest.predict(X_test)
```

```python
#printing scores
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

print('Train Accuracy:', forest.score(X_train, y_train))
print('Test Accuracy:', forest.score(X_test, y_test))
print("Recall Score:", metrics.recall_score(y_test,y_pred_forest))
```

The logistic regression model has been optimized to focus on achieving a higher recall score than accuracy because recall measures a model's ability to detect positive targets. In the context of this problem, this would mean focusing more on accurately identifying patients likely to contract heart disease and lowering the chance that the modeling would incorrectly categorize someone as unlikely to contract heart disease. This decision was also influenced by the fact that the data has less than 10% target outcomes, so a high accuracy is not neccessarily a good result if the percentage of unaccurate guesses are all patients with heart disease that the model said would not have it.

```python
# Train a Logistic Regression model and print the model score
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(max_iter = 5000, 
                            solver = 'saga', 
                            penalty = 'l2',
                            class_weight = {1:40})


logreg.fit(X_train, y_train)
y_pred_log = logreg.predict(X_test)
```
