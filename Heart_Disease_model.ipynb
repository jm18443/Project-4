```python
#import dependcies
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
#read in csv file and check data
df = pd.read_csv('C:/Users/tanne/Downloads/heart_2020_cleaned.csv/heart_2020_cleaned.csv')
df.head()
```

```python
#creating dummie variables
smoking =pd.get_dummies(df['Smoking'])
#dropint no column
smoking = smoking.drop(columns=["No"])
#renaming column
smoking = smoking.rename(columns = {'Yes':'Smoking'})

smoking.head()
```

```python
#creating dummie variables
stroke =pd.get_dummies(df['Stroke'])
#dropint no column
stroke = stroke.drop(columns=["No"])
#renaming column
stroke = stroke.rename(columns = {'Yes':'Stroke'})

stroke.head()
```

```python
#creating dummie variables
diffwalking =pd.get_dummies(df['DiffWalking'])
#dropint no column
diffwalking = diffwalking.drop(columns=["No"])
#renaming column
diffwalking = diffwalking.rename(columns = {'Yes':'DiffWalking'})

diffwalking.head()
```

```python
#creating dummie variables
diabetic =pd.get_dummies(df['Diabetic'])
#dropping no column
diabetic = diabetic.drop(columns=["No"])
#renameing columns
diabetic = diabetic.rename(columns = {'Yes':'Diabetic',
                                      'Yes (during pregnancy)':'Diabetic (duirng pregnancy)',
                                      'No, borderline diabetes':'Borderline Diabetic'})

diabetic.head(10)
```

```python
#creating dummie variables
physicalactivity =pd.get_dummies(df['PhysicalActivity'])
#dropint no column
physicalactivity = physicalactivity.drop(columns=["No"])
#renaming column
physicalactivity = physicalactivity.rename(columns = {'Yes':'PhysicalActivity'})

physicalactivity.head()
```

```python
#creating dummie variables
asthma =pd.get_dummies(df['Asthma'])
#dropint no column
asthma = asthma.drop(columns=["No"])
#renaming column
asthma = asthma.rename(columns = {'Yes':'Asthma'})

asthma.head()
```

```python
#creating dummie variables
kidneydisease =pd.get_dummies(df['KidneyDisease'])
#dropint no column
kidneydisease = kidneydisease.drop(columns=["No"])
#renaming column
kidneydisease = kidneydisease.rename(columns = {'Yes':'KidneyDisease'})


kidneydisease.head()
```

```python
#creating dummie variables
skincancer =pd.get_dummies(df['SkinCancer'])
#dropint no column
skincancer = skincancer.drop(columns=["No"])
#renaming column
skincancer = skincancer.rename(columns = {'Yes':'SkinCancer'})

skincancer.head()
```

```python
#creating dummie variables
sex =pd.get_dummies(df['Sex'])
#dropint no column
sex = sex.drop(columns=["Female"])

sex.head()
```

```python
import category_encoders
mapping = [
    {'col':'GenHealth',
    'mapping':{
        'Poor':1,
        'Fair':2,
        'Good':3,
        'Very good':4,
        'Excellent':5,
    }}
]
encoder = category_encoders.OrdinalEncoder(cols=['GenHealth'], return_df = True, mapping=mapping)

genhealth = encoder.fit_transform(df['GenHealth'])
genhealth
```

```python
race = pd.get_dummies(df['Race'])
race
```

```python
mapping2 = [
    {'col':'AgeCategory',
    'mapping':{
        '18-24':1,
        '25-29':2,
        '30-34':3,
        '35-39':4,
        '40-44':5,
        '45-49':6,
        '50-54':7,
        '55-59':8,
        '60-64':9,
        '65-69':10,
        '70-74':11,
        '75-79':12,
        '80 or older':13,     
    }}
]
encoder = category_encoders.OrdinalEncoder(cols=['AgeCategory'], return_df = True, mapping=mapping2)

age = encoder.fit_transform(df['AgeCategory'])
```

```python
df2 = df.drop(columns=['HeartDisease', 'Smoking', 'AlcoholDrinking', 'Stroke', 'DiffWalking', 'Sex',
                 'Race', 'Diabetic', 'PhysicalActivity', 'GenHealth','Asthma','KidneyDisease','SkinCancer','AgeCategory'])
```

```python
df2.head()
```

```python
dfs = [df2, smoking,heartdisease,alcoholdrinking ,stroke,diffwalking,
         physicalactivity ,diabetic ,asthma ,skincancer ,sex ,genhealth ,age ,race]
```

```python
dff = pd.concat(dfs, axis=1)
dff.head()
```

```python
list(dff)
```

```python
# Split the data into feature and target data
X = pd.DataFrame(dff.drop(['HeartDisease'], axis = 1))

y = pd.array(dff['HeartDisease'])
```

```python
# Split the data into X_train, X_test, y_train, y_test
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=444)
X_train
```

```python
# Train a Random Forest Classifier model and print the model score
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier()

forest.fit(X_train, y_train)
y_pred_forest = forest.predict(X_test)
```

```python
#printing scores
print('Train Accuracy:', forest.score(X_train, y_train))
print('Test Accuracy:', forest.score(X_test, y_test))
print("Recall Score:", metrics.recall_score(y_test,y_pred_forest))
```

```python
# Train a Logistic Regression model and print the model score
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()


logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
```

```python
#printing scores
print('Train Accuracy:', logreg.score(X_train, y_train))
print('Test Accuracy:', logreg.score(X_test, y_test))
print("Recall Score:", metrics.recall_score(y_test,y_pred))
```
